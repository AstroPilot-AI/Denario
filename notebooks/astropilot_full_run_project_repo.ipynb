# Set data description

```python
astro_pilot.set_data_description(r"""
A growing body of evidence indicates that transcranial ultrasound stimulation (TUS) may be a promising treatment for substance use disorders. A recent study has shown that TUS can induce alterations in brain connectivity and chemistry within the hour following stimulation. Importantly, the exceptional spatial precision and penetration depth of TUS allow researchers to modulate deep subcortical regions, such as the nucleus accumbens— a central hub of the reward network that governs addictive behaviours.

 

A number of US pilot trials have demonstrated that stimulation of the nucleus accumbens is safe and well tolerated in recently detoxified opioid-dependent individuals and that it reduces craving during a 90-day follow-up period. To date, no research group has explored the utility of TUS as an adjunct to existing talking therapies, taking advantage of its potential to influence brain connectivity and thus how individuals interpret and respond to therapy.

 

This project will be the first to explore the effects of TUS on enhancing the effectiveness of cognitive behavioural therapy (CBT) for the treatment of alcohol use disorder. 


The project team will work in partnership with government-funded addiction treatment providers in Manchester (UK) to recruit 120 individuals who have recently completed alcohol detoxification programmes and are receiving CBT to help reduce cravings and maintain abstinence upon returning to the community.

 

In this clinical trial, 60 participants will receive sham stimulation and 60 will receive genuine TUS. Individuals will attend eight combined CBT and TUS sessions over an eight-week period, during which they will receive their stimulation/shame protocol immediately before their standard CBT treatment. Participants will be followed up over a five-month period to compare the effects of sham versus real stimulation on alcohol use and craving. The project will be delivered over a four-year period, with a budget of approximately £600,000. This includes upfront costs of £200,000 for equipment purchases, £50,000 for consumables, £300,000 for staff costs, and £50,000 for participant expenses.


In this project, we will simulate the experimental data, to illustrate how such study could be carried out. 
"""
)
```
